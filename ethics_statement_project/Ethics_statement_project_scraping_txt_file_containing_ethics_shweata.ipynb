```python
import pandas as pd
import glob
import os
import pathlib

#HOME = "/Users/shweata/dictionary"
#PROJECT_DIR = os.path.join(HOME,"/e_cancer_clinical_trial_50")

# set filepath to search
path = '/Users/shweata/e_frontiers_500/**/*/*.xml'

# string to search for
search_term = "ethics"
# empty list to hold files that contain matching string
files_not_to_check = []
all_files = []
files_to_check = []

# looping through all the filenames returned
# returns only the fulltext.xml since we don't look in the sub-directories
# set recursive = False - doesn't look in sub-directories
for filename in glob.iglob(path, recursive=False):
    # adding error handling just in case!
    try:
        with open(filename) as f:
            # read the file as a string
            contents = f.read()
            # if the search term is found append to the list of files
            if(search_term in contents):
                files_not_to_check.append(filename)
    except:
        pass


# looping through all the filenames returned
# returns all the fulltext and sections which has the search_term   
# set recursive = True to look in sub-directories too
for filename in glob.iglob(path, recursive=True):
    try:
        with open(filename) as f:
            # read the file as a string
            contents = f.read()
            # if the search term is found append to the list of files
            if(search_term in contents):
                all_files.append(filename)
    except:
        pass

# Currently, I don't know how to not include full-text when I glob. I create another list with no full-text called files_to_check here. 
files_to_check = [x for x in all_files if x not in files_not_to_check]
files_to_check
```

```python
# Named entity-recognition 
nlp = spacy.load("en_core_web_sm")


#!python -m pip install icecream
#from icecream import ic
doc = nlp(text)


entities = []
labels = []
position_start = []
position_end = []

for ent in doc.ents:
    entities.append(ent)
    labels.append(ent.label_)
    position_start.append(ent.start_char)
    position_end.append(ent.end_char)
    
df = pd.DataFrame({'Entities':entities,'Labels':labels,'Position_Start':position_start, 'Position_End':position_end})
#pd.set_option("display.max_rows", None, "display.max_columns", None)
df[df["Labels"] == 'ORG']
df.to_csv('entity_recognition_spacy_cancer_trial_50.csv', encoding='utf-8')
#ic(entities, labels)
df
```

```python
# Named entity recognition using nltk - Words
import nltk
words = nltk.word_tokenize(text)
pos_tags = nltk.pos_tag(words)
#tokenize to words
words = nltk.word_tokenize(text)
chunks = nltk.ne_chunk(pos_tags, binary=False) #either NE or not NE
    
entities =[]
labels =[]
for chunk in chunks:
    if hasattr(chunk,'label'):
        #print(chunk)
        entities.append(' '.join(c[0] for c in chunk))
        labels.append(chunk.label())
        
entities_labels = list(set(zip(entities, labels)))
entities_df = pd.DataFrame(entities_labels)
entities_df.columns = ["Entities","Labels"]
pd.set_option("display.max_rows", None, "display.max_columns", None)
entities_df[entities_df["Labels"] == 'ORGANIZATION']
entities_df
```

```python
# Named-Entity Recogntion using nltkk - sentences
entities = []
labels = []

sentence = nltk.sent_tokenize(text)
for sent in sentence:
    for chunk in nltk.ne_chunk(nltk.pos_tag(nltk.word_tokenize(sent)),binary=False):
        if hasattr(chunk,'label'):
            entities.append(' '.join(c[0] for c in chunk))
            labels.append(chunk.label())
            
entities_labels = list(set(zip(entities,labels)))

entities_df = pd.DataFrame(entities_labels)
entities_df.columns = ["Entities","Labels"]
entities_df[entities_df["Labels"] == 'ORGANIZATION']
entities_df
```
